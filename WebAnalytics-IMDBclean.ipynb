```python
# Generate the binary list for the genres of the movie table
genres = ['Action','Drama','Thriller','Crime','Adventure','Comedy','Romance',
          'Sci-Fi','Fantasy','Horror','Western','War','Mystery','History','Family',
          'Animation','Sport','Music','Biography','Musical']
genreL = list(movie_f.movieGenre)
# Action
Action = []
for i in genreL:
    if genres[0] in i:
        Action.append(1)
    else:
        Action.append(0)
# Drama
Drama = []
for i in genreL:
    if genres[1] in i:
        Drama.append(1)
    else:
        Drama.append(0)
# Thriller
Thriller = []
for i in genreL:
    if genres[2] in i:
        Thriller.append(1)
    else:
        Thriller.append(0)
# Crime
Crime = []
for i in genreL:
    if genres[3] in i:
        Crime.append(1)
    else:
        Crime.append(0)
# Adventure
Adventure = []
for i in genreL:
    if genres[4] in i:
        Adventure.append(1)
    else:
        Adventure.append(0)
# Comedy
Comedy = []
for i in genreL:
    if genres[5] in i:
        Comedy.append(1)
    else:
        Comedy.append(0)
# Romance
Romance = []
for i in genreL:
    if genres[6] in i:
        Romance.append(1)
    else:
        Romance.append(0)
# Sci-Fi
Sci_Fi = []
for i in genreL:
    if genres[7] in i:
        Sci_Fi.append(1)
    else:
        Sci_Fi.append(0)
# Fantasy
Fantasy = []
for i in genreL:
    if genres[8] in i:
        Fantasy.append(1)
    else:
        Fantasy.append(0)
# Horror
Horror = []
for i in genreL:
    if genres[9] in i:
        Horror.append(1)
    else:
        Horror.append(0)
# Western
Western = []
for i in genreL:
    if genres[10] in i:
        Western.append(1)
    else:
        Western.append(0)
# War
War = []
for i in genreL:
    if genres[11] in i:
        War.append(1)
    else:
        War.append(0)
# Mystery
Mystery = []
for i in genreL:
    if genres[12] in i:
        Mystery.append(1)
    else:
        Mystery.append(0)
# History
History = []
for i in genreL:
    if genres[13] in i:
        History.append(1)
    else:
        History.append(0)
# Family
Family = []
for i in genreL:
    if genres[14] in i:
        Family.append(1)
    else:
        Family.append(0)
# Animation
Animation = []
for i in genreL:
    if genres[15] in i:
        Animation.append(1)
    else:
        Animation.append(0)
# Sport
Sport = []
for i in genreL:
    if genres[16] in i:
        Sport.append(1)
    else:
        Sport.append(0)
# Biography
Biography = []
for i in genreL:
    if genres[18] in i:
        Biography.append(1)
    else:
        Biography.append(0)
# Musical
Musical = []
for i in genreL:
    if genres[19] in i:
        Musical.append(1)
    else:
        Musical.append(0)
```
