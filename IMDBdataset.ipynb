# Crawl Data From IMDb Website

## Get basic information from IMDb Website

```python
from bs4 import BeautifulSoup 
import urllib.request
```

```python

url = 'https://www.imdb.com/search/title/?title_type=feature&genres=action&explore=genres&ref_=adv_explore_rhs'
page = urllib.request.urlopen(url)
soup_l = BeautifulSoup(page,'html.parser')
genre_l = soup_l.find('div',{'class':'aux-content-widget-2'}).find_all('td')
genre = []
for i in genre_l:
    genre.append(i.find('a').get_text())
```

```python
genre[:20] # check
```

```python
g1 = genre[:16] # need to crawl 1000
g2 = genre[-9:-5] # need to crawl all 
```

## Movie genres that need to crawl 1000 movies

```python
# Genres need to crawl 1000
p = [i for i in range(1000)]
page_A = p[1::50]
url_A = []
for g in g1:
    for i in page_A:
        url_A.append('https://www.imdb.com/search/title/?title_type=feature&genres='+g+'&sort=num_votes,desc&start='+str(i)+'&explore=genres')

main_A = [] # The list of all of the movies
n = 1
for i in url_A:
    SourcePage = urllib.request.urlopen(i)
    soup = BeautifulSoup(SourcePage,'html.parser')
    main_A.append(soup.find('div',{'class':'lister-list'}))
    print(n)
    n += 1
```

```python
# check
print(len(main_A))
```

```python
movieName_A = []
movieUrl_A = []
movieCerti_A = []
movieRuntime_A = []
movieGenre_A = []
movieOR_A = []
n = 1
for p in main_A: 
    movies = p.find_all('h3',{'class':'lister-item-header'})# Movie titles and links
    con = p.find_all('div',{'class':'lister-item-content'})# The content part
    print(n)
    n += 1
    for i in movies:
        movieName_A.append(i.find('a').get_text())# The title
        movieUrl_A.append(i.find('a').get('href'))# The link
    for r in con:
        if r.find('span',{'class':'certificate'})!=None:# Certification: PR-13
            movieCerti_A.append(r.find('span',{'class':'certificate'}).get_text())
        else:
            movieCerti_A.append('Null')
        if r.find('span',{'class':'runtime'})!=None:# The length of the movie
            movieRuntime_A.append(r.find('span',{'class':'runtime'}).get_text())
        else:
            movieRuntime_A.append('Null')
        if r.find('span',{'class':'genre'})!=None:# Genre
            movieGenre_A.append(r.find('span',{'class':'genre'}).get_text())
        else:
            movieGenre_A.append('Null')
        if r.find('strong')!=None:# Ratings
            movieOR_A.append(r.find('strong').get_text())
        else:
            movieOR_A.append('Null')
```

```python
len(set(movieCerti_A))
import pandas as pd
mA = {}
mA['url'] = movieUrl_A
mA['movieName'] = movieName_A
mA['movieCertification'] = movieCerti_A
mA['movieRuntime'] = movieRuntime_A
mA['movieGenre'] = movieGenre_A
mA['movieOveralRating'] = movieOR_A
movieinfoA = pd.DataFrame(mA)
movieA = movieinfoA.drop_duplicates(subset='url')
# movie = mo.set_index(['url','movieName'])
# movie.sort_index() # Check for the repeating values
```

```python
movieUrlA = list(movieA.url)
movieNameA = list(movieA.movieName)
movieCertiA = list(movieA.movieCertification)
movieRuntimeA = list(movieA.movieRuntime)
movieGenreA = list(movieA.movieGenre)
movieORA = list(movieA.movieOveralRating)
len(movieUrlA) # check
```

```python
# check the results: usualy the length of these dataets will be 1:1:1:1
print(len(movieNameA),'\n',len(movieCertiA),'\n',len(movieRuntimeA),'\n',len(movieGenreA),'\n',len(movieORA))
```

```python
reviewUrlA = []# The url of the reviews
for i in movieUrlA:
    p = 'https://www.imdb.com' + str(i) + 'reviews?ref_=tt_ov_rt'
    reviewUrlA.append(p)
```
